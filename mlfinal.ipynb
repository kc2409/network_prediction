```python
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder


#df_T = pd.get_dummies(df_train, columns = ["protocol_type","service","flag","type_of_attack"],prefix="type")

#df_T
le=LabelEncoder()
df_train['protocol_type']= le.fit_transform(df_train['protocol_type'])
df_test['protocol_type']= le.transform(df_test['protocol_type'])

d_T=df_train.append(df_test)
d_T['flag']= le.fit(d_T['flag'].astype(str))
df_train['flag']= le.transform(df_train['flag'].astype(str))
df_test['flag']= le.transform(df_test['flag'].astype(str))

lmap={"ipsweep.":0,"satan.":1,"portsweep.":2,"back.":3,"normal.":4}
df_train["type_of_attack"]=df_train["type_of_attack"].map(lmap)

df_train['service']=le.fit_transform(df_train['service'])
df_test.loc[df_test["service"]=="aol","service"]="other"
df_test.loc[df_test["service"]=="harvest","service"]="other"
df_test['service']=le.transform(df_test['service'])

#df_train['type_of_attack']= le.fit_transform(df_train['type_of_attack'])



  
#scaler = StandardScaler()
#scaled = scaler.fit_transform(df_train)
#scalar.fit(df_train)
#scaled_data = scalar.transform(df_train)
#df_train = pd.DataFrame(scaled)
df_train


```

```python
Y=df_train["type_of_attack"]
x=df_train.drop(["type_of_attack"],axis=1)
from sklearn.linear_model import LogisticRegression
lgr = LogisticRegression(max_iter = 1100000,class_weight="balanced")
lgr.fit(x,Y)
print("Accuracy",(lgr.score(x,Y)))
```

```python
Id= df_test['Id']
pred2 = lgr.predict(df_test.astype(np.float))
prd2=[]
lmap={0:"ipsweep.",1:"satan.",2:"portsweep.",3:"back.",4:"normal."}
for i in range(len(pred2)):
    prd2.append(lmap[int(pred2[i])])
print(prd2[:20])
dict1 = {'Id': Id, 'type_of_attack': prd2} 
     
df = pd.DataFrame(dict1)
  
# saving the dataframe
df.to_csv('file1.csv',index=False)
```
